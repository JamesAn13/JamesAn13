```python
from moviepy.editor import VideoFileClip, concatenate_videoclips
```

```python
def make_cilp_video(path, save_path, start_t, end_t):
    clip = VideoFileClip(path).subclip(start_t, end_t)
    clip.write_videofile(save_path)
```
