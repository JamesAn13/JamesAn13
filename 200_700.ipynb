```python
import os
import cv2
from ultralytics import YOLO
import mediapipe as mp

# YOLO 모델 로드
yolo_model = YOLO('yolov5s.pt')  # YOLOv5 사전 학습 모델

# MediaPipe 초기화
mp_pose = mp.solutions.pose
mp_face_mesh = mp.solutions.face_mesh

pose = mp_pose.Pose(static_image_mode=False, min_detection_confidence=0.5, min_tracking_confidence=0.5)
face_mesh = mp_face_mesh.FaceMesh(static_image_mode=False, max_num_faces=1, min_detection_confidence=0.5)

# 내 영상 데이터 경로
video_folder = '/content/drive/MyDrive/data290/'
# 결과 저장 경로
output_folder = '/content/drive/MyDrive/processed_videos2/'
os.makedirs(output_folder, exist_ok=True)

def process_video(video_path, output_path):
    cap = cv2.VideoCapture(video_path)
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')

    # 출력 비디오 설정
    out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # YOLO로 사람 객체 검출
        results = yolo_model(frame)
        for result in results:
            for box in result.boxes.xyxy.tolist():  # Bounding box 좌표
                x1, y1, x2, y2 = map(int, box[:4])  # 박스 좌표
                label = int(result.boxes.cls[0])   # 클래스 ID (0: person)

                # 클래스가 'person'인 경우에만 처리
                if label == 0:
                    # Bounding box 그리기
                    cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 0, 255), 2)  # 빨간색 테두리
                    cv2.putText(frame, "Person", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)

                    # 사람 영역 크롭 후 MediaPipe 처리
                    cropped_frame = frame[y1:y2, x1:x2]

                    # MediaPipe Pose (몸 랜드마크)
                    mp_pose_results = pose.process(cv2.cvtColor(cropped_frame, cv2.COLOR_BGR2RGB))
                    if mp_pose_results.pose_landmarks:
                        for lm in mp_pose_results.pose_landmarks.landmark:
                            # 랜드마크 좌표 변환 (Bounding box 기준 → 전체 프레임 기준)
                            cx = int(x1 + lm.x * (x2 - x1))
                            cy = int(y1 + lm.y * (y2 - y1))
                            cv2.circle(frame, (cx, cy), 3, (0, 255, 0), -1)  # 녹색 점 표시

                    # MediaPipe Face Mesh (얼굴 랜드마크)
                    mp_face_results = face_mesh.process(cv2.cvtColor(cropped_frame, cv2.COLOR_BGR2RGB))
                    if mp_face_results.multi_face_landmarks:
                        for face_landmarks in mp_face_results.multi_face_landmarks:
                            for lm in face_landmarks.landmark:
                                # 랜드마크 좌표 변환
                                cx = int(x1 + lm.x * (x2 - x1))
                                cy = int(y1 + lm.y * (y2 - y1))
                                cv2.circle(frame, (cx, cy), 1, (255, 0, 0), -1)  # 파란색 점 표시

        # 처리된 프레임 저장
        out.write(frame)

    cap.release()
    out.release()

# 영상 처리: 반만 선택
video_files = [f for f in os.listdir(video_folder) if f.endswith(('.mp4', '.avi', '.mkv'))]
half_files = video_files[:len(video_files) // 2]  # 절반 선택

for video_name in half_files:
    input_path = os.path.join(video_folder, video_name)
    output_path = os.path.join(output_folder, f'processed_{video_name}')
    process_video(input_path, output_path)
    print(f'{video_name} 처리 완료!')
```

```python
import os
import cv2
from ultralytics import YOLO
import mediapipe as mp

# YOLO 모델 로드
yolo_model = YOLO('yolov5s.pt')  # YOLOv5 사전 학습 모델

# MediaPipe Pose 초기화
mp_pose = mp.solutions.pose
pose = mp_pose.Pose(static_image_mode=False, min_detection_confidence=0.5, min_tracking_confidence=0.5)

# 내 영상 데이터 경로
video_folder = '/content/drive/MyDrive/data290/'
output_folder = '/content/drive/MyDrive/processed_videos_yoloFirst/'
os.makedirs(output_folder, exist_ok=True)

# Pose 랜드마크 연결 정보
POSE_CONNECTIONS = [
    (11, 13), (13, 15),  # 왼팔
    (12, 14), (14, 16),  # 오른팔
    (11, 12),            # 어깨 연결
    (11, 23), (12, 24),  # 몸통 연결
    (23, 24),            # 골반 연결
    (23, 25), (25, 27),  # 왼다리
    (24, 26), (26, 28)   # 오른다리
]

def process_video(video_path, output_path):
    cap = cv2.VideoCapture(video_path)
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')

    # 출력 비디오 설정
    out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # YOLO로 사람 객체 검출
        results = yolo_model(frame)
        person_boxes = []

        # 사람 객체만 추출 (label == 0: 사람)
        for result in results:
            for box, conf, label in zip(result.boxes.xyxy.tolist(), result.boxes.conf.tolist(), result.boxes.cls.tolist()):
                if label == 0:  # 사람 객체만
                    person_boxes.append((box, conf))  # 사람의 bounding box와 confidence 저장

        if person_boxes:
            # 가장 높은 확신도(confidence)를 가진 사람 객체를 선택
            person_boxes.sort(key=lambda x: x[1], reverse=True)  # confidence 기준으로 내림차순 정렬
            best_box = person_boxes[0][0]  # 가장 높은 confidence를 가진 box

            x1, y1, x2, y2 = map(int, best_box)

            # YOLO로 검출된 사람 객체에 빨간색 테두리 그리기
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 0, 255), 2)  # 빨간색 테두리

            # YOLO로 검출된 사람 객체에 대해 MediaPipe Pose 처리
            cropped_frame = frame[y1:y2, x1:x2]
            mp_results = pose.process(cv2.cvtColor(cropped_frame, cv2.COLOR_BGR2RGB))
            if mp_results.pose_landmarks:
                for connection in POSE_CONNECTIONS:
                    start_idx, end_idx = connection
                    start = mp_results.pose_landmarks.landmark[start_idx]
                    end = mp_results.pose_landmarks.landmark[end_idx]

                    # 좌표 계산 (Bounding box 기준 → 전체 프레임 기준 변환)
                    start_x, start_y = int(x1 + start.x * (x2 - x1)), int(y1 + start.y * (y2 - y1))
                    end_x, end_y = int(x1 + end.x * (x2 - x1)), int(y1 + end.y * (y2 - y1))

                    # 선 그리기
                    cv2.line(frame, (start_x, start_y), (end_x, end_y), (0, 255, 0), 3)  # 초록색 선
                    cv2.circle(frame, (start_x, start_y), 5, (0, 255, 0), -1)  # 점 표시
                    cv2.circle(frame, (end_x, end_y), 5, (0, 255, 0), -1)

        # 처리된 프레임 저장
        out.write(frame)

    cap.release()
    out.release()

# 영상 처리 반복
video_files = [f for f in os.listdir(video_folder) if f.endswith(('.mp4', '.avi', '.mkv'))][:5]  # 최대 10개만 선택
for video_name in video_files:
    input_path = os.path.join(video_folder, video_name)
    output_path = os.path.join(output_folder, f'processed_{video_name}')
    process_video(input_path, output_path)
    print(f'{video_name} 처리 완료!')
```
